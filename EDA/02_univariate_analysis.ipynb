# Univariate Analysis
  Analysis on a single column


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv')
df.head(5)
```

## 1. Categorical Data


> (a) ***Countplot***
---

if we have categorical data, it is always beneficial to use the count plot to gain some insights initially.



```python
sns.countplot(x = 'Survived', data = df)
```

```python
sns.countplot(x = 'Pclass', data = df)
```

```python
sns.countplot(x = 'Sex', data = df)
```

```python
sns.countplot(x = 'Embarked', data = df)
```

> (b) Pie Chart:
---

```python
sizes = df['Survived'].value_counts()
plt.pie(sizes, labels = ['Not Survived', 'Survived'], autopct='%1.1f%%')
plt.title('Survived vs Not Survived')
plt.show()
```

```python
plt.pie(df['Pclass'].value_counts(), labels= ['3', '1', '2'], autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(df['Sex'].value_counts(), labels= ['M', 'F'], autopct='%1.1f%%')
plt.show()
```
