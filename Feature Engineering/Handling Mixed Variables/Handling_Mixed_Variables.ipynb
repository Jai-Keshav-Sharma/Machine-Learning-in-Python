```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day33-handling-mixed-variables/titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
# extract numerical part
# It converts the values in the column df['number'] to numeric integers, handling any non-numeric values safely by converting them to NaN.
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
# extract categorical part
# Copies values from df['number'] only if the corresponding number_numerical value is NaN.
# Otherwise, it puts NaN in the new column.

df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
# HANDLING CABIN
df['Cabin_num'] = df['Cabin'].str.extract('(\d+)') # CAPTURES NUMERICAL PART
df['cabin_cat'] = df['Cabin'].str[0] # CAPTURES THE FIRST CHARACTER

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# EXTRACT THE LAST BIT OF TICKET AS NUMBER
# s.split() splits the string s by spaces.
# [-1] takes the last part of that split.

df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')

# EXTRACT THE FIRST BIT OF TICKET AS CATEGORY

df['ticket_cat'] = df['Ticket'].apply(lambda s:s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])

df.head(20)
```
