```python
pd.get_dummies(df, columns=['fuel', 'owner'], drop_first=True)
```

# 3. OneHot Encoding Using Sklearn

##### Using sklearn, we first separate the columns that we want to encode, then after encoding, we again append them to the dataframe

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:, 0:4], df.iloc[:, -1], test_size=0.2, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import OneHotEncoder

# sparse_output = FALSE means the output will not be a sparse matrix
# drop = 'first' ensures removing the first dummy column

ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)
```

```python
# separating the fuel and owner columns and then applying One Hot Encoding
X_train_new = ohe.fit_transform(X_train[['fuel', 'owner']])
X_test_new = ohe.transform(X_test[['fuel', 'owner']])
```

```python
X_train_new
```

```python
# joining after OneHot Encoding
np.hstack((X_train[['brand', 'km_driven']].values, X_train_new))
```

# 4. OneHot Encoding with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
# threshold = 100 -> brands below 100 cars will be stored in a single separate column
threshold = 100
```

```python
# brands having less cars then threshold
repl = counts[counts <= threshold].index
```

```python
# categories having more than 100 cars
# 'uncommon' has brands having cars  <= 100

pd.get_dummies(df['brand'].replace(repl, 'uncommon'))
```
