```python
import numpy as np
import pandas as pd
```

```python
date = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day34-handling-date-and-time/orders.csv')
time = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day34-handling-date-and-time/messages.csv')
```

```python
# DATE IS AN OBJECT BY DEFAULT IN PANDAS

date.info()
```

```python
# CONVERTING TO DATE-TIME DATA TYPE

date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

# Working With Dates

### 1. Extracting Year

```python
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

### 2. Extracting Month


```python
date['date_month'] = date['date'].dt.month
```

```python
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
```

```python
date.sample(4)
```

### 3. Extracting Days

```python
date['date_day'] = date['date'].dt.day
```

```python
date.head()
```

```python
date['date_day_of_week'] = date['date'].dt.dayofweek
```

```python
date.head()
```

```python
date['date_day_name'] = date['date'].dt.day_name()
```

```python
date.head()
```

```python
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
# IS WEEKEND ?
date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Sunday', 'Saturday']), 1, 0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

```python
date['week_of_year'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

### 4. Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

### 5. Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1, 2]), 1, 2)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

### 6. Extracting Time Elapsed Between Dates


```python
import datetime
```

```python
today = datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

months_passed = (today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month)
months_passed
```

# Working With Time

```python
time.head()
```

```python
# DATE IS AN OBJECT
time.info()
```

```python
time['date'] = pd.to_datetime(time['date'])
```

```python
time.info()
```

### 1. Extracting Hour, Minute and Second

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

### 2. Extracting the Time Part

```python
time['time'] = time['date'].dt.time

time.head()
```

### 3. Time Difference

```python
today - time['date']
```

```python
# IN SECONDS

seconds_passed = (today - time['date']) / np.timedelta64(1, 's')
seconds_passed
```

```python
# IN HOURS

hours_passed = (today - time['date']) / np.timedelta64(1, 'h')
hours_passed
```
